```python
import json
import matplotlib.pyplot as plt
import requests
import time

data_list = []

while len(data_list) < 105:
    x = requests.get("http://api.open-notify.org/iss-now.json")
    
    if x.status_code == 200:
        data = x.json()
        timestamp = data["timestamp"]
        latitude = float(data["iss_position"]["latitude"])
        longitude = float(data["iss_position"]["longitude"])
    
        data_list.append((timestamp, latitude, longitude))
        time.sleep(10)
        
times = [point[0] for point in data_list]
lats = [point[1] for point in data_list]
longs = [point[2] for point in data_list]

plt.plot(longs, lats)

plt.title("International Space Station Location")
plt.xlabel("Longitude")
plt.ylabel("Latitude")

plt.grid(True)
plt.show()
```
